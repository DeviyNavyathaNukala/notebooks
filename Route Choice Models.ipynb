```python
N=7
```

```python
def create_graph(N):
    g = nx.Graph()
    g.add_nodes_from(map(str,range(N*N)))

    for i in range(N):    
        for j in range(N):
            if j<N-1: g.add_edge(str(i*N+j), str(i*N+j+1))
            if i<N-1: g.add_edge(str(i*N+j), str(i*N+j+N))
    return g
```

```python
g = create_graph(N)
nx.draw_spectral(g)
```

```python
def find_path(origin,graph, dest):

    remaining = [iter(graph[origin])]
    visited = [origin]
    while remaining:

        current = remaining[-1]
        child = next(current,None)

        if not child:
            visited.pop()
            remaining.pop()
        elif child == dest:
            yield 1 #visited + [child]
        elif not (child in visited):
            remaining.append(iter(graph[child]))
            visited.append(child)
```

```python
N = 6
g = create_graph(N)
gen = find_path('0',g,str(N*N-1))
```
