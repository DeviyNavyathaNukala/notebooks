```python
class Reaction:
    
    def __init__(self, equation_number,loss,production, rate):
        self.equation_number = equation_number
        self.loss = loss
        self.production = production
        self.rate = rate
    
    def __repr__(self):
        return str([self.equation_number,self.loss,self.production,self.rate])
```

```python
with open('/Users/alehmann/projects/DASK/EMEP_DEHM.mch','rt') as f:
    data = f.readlines()
    
    
reactions = []
for l in data:
    if l.startswith('!') or l == '\n' or l.startswith('END'): continue
        
    equat = l.split('->')
    
    if len(equat) == 2:
        lhs, rhs = equat
        equation = lhs.split()[0]
        loss_species = lhs.split()[1:]
        prod_species = rhs.split()
    elif l.startswith('&'):
        prod_species.append(l.split()[1:])
    else:
        reactions.append(Reaction(equation,loss_species,prod_species,l.strip()))
    
        print reactions[-1]

```

```python
def thermal(A,E_R):
    """The reaction rate for thermal reactions
        T is the temperature in Kelvin)"""
    return lambda T: A*exp(E_R/T)
```
