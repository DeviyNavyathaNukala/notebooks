```python
import codecs

root = '/Users/alehmann/Documents/Documents/HIH/BDA/BDA2017/data/'
def read_energi_data(f):
    with codecs.open(root+f,encoding='8859') as f:
        col = []
        l=f.readline().strip()
        while l:
            col.append(' '.join(l.split()[1:]))
            l=f.readline().strip()
        
        dat = pd.read_csv(f,sep=';', skipinitialspace=True,lineterminator ='\n',infer_datetime_format=True)
        dat = dat.drop(dat.columns[-1],1)
        
        dat.columns = [dat.columns[0]] + col
        dat = dat.set_index(dat.columns[0])
    return dat

```

```python
from sklearn.model_selection import GridSearchCV,cross_validate
from sklearn.pipeline import Pipeline
```

```python
from sklearn.model_selection import train_test_split
```

```python
energi_train,energi_test,co2_train,co2_test = train_test_split(energi,co2,test_size=0.3)
```

```python
pipe = Pipeline([('pca',PCA(n_components=5)),('svr',SVR(kernel='linear'))])
pipe.fit(energi_train,co2_train)
```

```python
plot(pipe.predict(energi_test))
co2_test.plot()
```

```python
param ={'svr__C':[1,10,100],'svr__epsilon':[0.001,0.01,0.1]}
tune = GridSearchCV(pipe,param,n_jobs=-1)
tune.estimator.get_params().keys()
```

```python
tune.fit(energi,co2)
```
