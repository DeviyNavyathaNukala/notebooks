```python
def BPR(x,t=50.,C=10000.,alfa=0.5,beta=4):
    return t*(1+ alfa*(x/C)**beta)
```

```python
x=linspace(0,19000,num=100)
plot(x,BPR(x))
```

```python
def traveltime(x,max=19000):
    return (x*BPR(x)+(max-x)*BPR(max-x))/max
```

```python
plot(x,traveltime(x))
```
