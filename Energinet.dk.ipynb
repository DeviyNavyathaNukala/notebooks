```python
import codecs
def read_energi_data(f):
    with codecs.open(root+f,encoding='8859') as f:
        col = []
        l=f.readline().strip()
        while l:
            col.append(' '.join(l.split()[1:]))
            l=f.readline().strip()
        
        dat = pandas.read_csv(f,sep=';', skipinitialspace=True,lineterminator ='\n',infer_datetime_format=True)
        dat = dat.drop(dat.columns[-1],1)
        
        dat.columns = [dat.columns[0]] + col
        dat = dat.set_index(dat.columns[0])
    return dat

df = []
for d in os.listdir(root)[2:]:
    if os.path.isdir(root+d):
        print d
        continue
    df.append(read_energi_data(d))
data = pandas.concat(df) 

```

```python
CO2 = data[u'CO2 udledning']
CO2 = CO2.fillna(0)
r_co2 =  CO2[CO2>0]
s = pandas.rolling_median(r_co2,36)
ss = pandas.rolling_median(r_co2,288)
s = pandas.rolling_mean(s,3*24*60/5)
#figure.figsize = 250,15
ss.plot()
figure()
p = s.plot(title='CO$_2$')
#p.figure.set_figwidth(25)
#p.figure.set_figheight(15)
ylabel('(g/kwh)',fontsize='large')
xlabel('Date')
p.set_xticklabels(p.xaxis.get_majorticklabels(), rotation=45)
savefig('/Users/alehmann/Documents/Documents/HIH/phd_application/papers/dissertation/figures/CO2_2_new.png')
```
